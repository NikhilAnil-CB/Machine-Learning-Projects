```python
X=df[['Weight']]
y = df['Height']
```

```python
X_Series = df['Weight']
np.array(X_Series).shape
```

```python
np.array(y).shape
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
X_train,y_train,X_test,y_test = train_test_split(X,y,test_size= 0.25, random_state= 42)
```

```python
#Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
```

```python

# Reshape the Series into a 2D array
X_test = X_test.values.reshape(-1, 1)
X_test = scaler.transform(X_test)
```

```python
X_test
```

```python
#Applying Linear Regression Model
from sklearn.linear_model import LinearRegression
lr = LinearRegression(n_jobs=-1)
```

```python
lr.fit(X_train,y_train)
```

```python
min_samples = min(len(X_train), len(y_train))
X_train = X_train[:min_samples]
y_train = y_train[:min_samples]
```

```python
lr.fit(X_train,y_train)
```

```python
print("Coefficient or slope:",lr.coef_)
print("Intercept:",lr.intercept_)
```

```python
predictions = lr.predict(X_test)
predictions = predictions.reshape(-1)  # Flatten predictions if needed
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_test, predictions, color='red')

```
